```python
y=df['output']
x=df.drop('output',axis=1)# splitting to input and output columns
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=0)#splitting to train and test datasets
```

```python
# Normalizing data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### Logistic regression

```python
#Logistic regression model
warnings.filterwarnings('ignore')
lg = LogisticRegression()
lg.fit(x_train, y_train)
ypred=lg.predict(x_test)
print(accuracy_score(y_test, ypred)*100,"%")
```

### Random Forest

```python
# Random forest model
rf = RandomForestRegressor(n_estimators = 100, random_state = 0)  
rf.fit(x_train, y_train)  
predicted = rf.predict(x_test)
print(accuracy_score(y_test, predicted.round())*100, "%")
```

### Support vector machine

```python
#support vector machine model
svm = SVC()
svm.fit(x_train, y_train)  
ypred1 = svm.predict(x_test)
print("The accuracy of SVM is : ", accuracy_score(y_test, ypred1)*100, "%")
```

### K nearest neighbors

```python
# knn model
knn = KNeighborsClassifier(n_neighbors = 7)  
knn.fit(x_train, y_train)
predictedk = knn.predict(x_test)
print(accuracy_score(y_test, predictedk.round())*100, "%")
```

### Naive bayes

```python
# naive bayes model
nb = GaussianNB()
nb.fit(x_train, y_train)
predictedn = nb.predict(x_test) 
print(accuracy_score(y_test, predictedn)*100, "%")
```

## Conclusion

SVM is best performing model with 93.44% accuracy
